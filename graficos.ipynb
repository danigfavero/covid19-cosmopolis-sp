```python
from datetime import datetime
import csv

atributos = [dict() for i in range(12)]
monitoramento, investigados, ag_resultados, descartados, confirmados, curados, \
isolamento, obitos, obitos_inv, leitos_covid, uti_covid, uti_nao_covid = atributos

with open('dados.csv', newline='') as arquivo:
    leitor = csv.reader(arquivo)
    for linha in leitor:
        data = datetime.strptime(linha[0], '%d-%m-%Y').date()
        for i in range(12):
            if linha[i+1] != "":
                atributos[i][data] = int(linha[i+1])
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

x = monitoramento.keys()
y = monitoramento.values()

plt.plot(x, y)
plt.xlabel('Tempo')
plt.ylabel('Nº')
plt.title('Monitoramento')
plt.gcf().autofmt_xdate()
plt.show()
```
